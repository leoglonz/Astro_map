```python
import pynbody
import pandas as pd
import numpy
import pickle
```

```python
# full path for elena h329 simulation.
path = '/Users/leoglonz/Desktop/Stellar_Feedback_Analysis/SNeData/timesteps_data/h329.data'

# unpacking all list data into objs and then converting to df.
objs = []
f = open(path, 'rb')
while 1: 
    try: 
        objs.append(pickle.load(f)) 
        
    except EOFError: 
        break 
        
f.close()

df = pd.DataFrame(objs)
```
